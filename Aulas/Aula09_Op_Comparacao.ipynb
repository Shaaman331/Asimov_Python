```python
print("Operadores de comparação em cadeia")
print("Igual a : ", 5 == 5)
5 == 5  # True
"hello" == "hello"  # True
5 == 10  # False

print("Diferente de : ", 5 != 5)
5 != 10  # True
"hello" != "world"  # True
5 != 5  # False

print("Maior que")
10 > 5  # True
5 > 10  # False

print("Menor que")
5 < 10  # True
10 < 5  # False

print("Maior ou igual")
10 >= 10  # True
15 >= 10  # True
5 >= 10  # False

print("Menor ou igual")
5 <= 5  # True
5 <= 10  # True
15 <= 10  # False

print("Exemplos praticos")
a = 10
b = 20

print(a == b)  # False
print(a != b)  # True
print(a > b)   # False
print(a < b)   # True
print(a >= b)  # False
print(a <= b)  # True


print("Exemplo com strings")
print("apple" < "banana")  # True (ordem alfabética)
print("apple" == "apple")  # True

print("Exemplo com listas")
print([1, 2, 3] == [1, 2, 3])  # True
print([1, 2, 3] != [4, 5, 6])  # True


print("Operador and")
idade = 25

if idade >= 18 and idade <= 65:
    print("A idade está dentro do intervalo de 18 a 65 anos.")
else:
    print("A idade está fora do intervalo.")
    

print("Combinando mais de uma condição")
x = 10
y = 20
z = 30

if x < y and y < z and z > x:
    print("Todas as condições são verdadeiras.")



print("Operador or")

numero = 150

if numero < 0 or numero > 100:
    print("O número está fora do intervalo de 0 a 100.")
else:
    print("O número está dentro do intervalo.")


print("Verificando multiplas condições")
dia = "sábado"

if dia == "sábado" or dia == "domingo":
    print("É fim de semana!")
else:
    print("É dia útil.") 


print("Combinando mais de duas condições")
cor = "vermelho"

if cor == "vermelho" or cor == "azul" or cor == "verde":
    print("A cor é primária ou secundária.")
else:
    print("A cor não é primária nem secundária.")
```
